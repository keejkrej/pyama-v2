# Pyama crop

Edit the **Config** cell, then run the crop cells.

Requires `bbox/Pos*.csv` from the desktop Align workflow. Writes ROI stacks under `WORKSPACE/roi/`.
After cropping, continue in `analyze.ipynb`.

## Config

Set **workspace**, **source**, and **positions** to crop. `FORCE_CROP` rewrites existing ROI dirs.

```python
from pathlib import Path

# Dataset root (bbox/, roi/, …)
WORKSPACE = Path(r"C:\path\to\dataset")

# ND2 or CZI source (requires bbox/Pos*.csv from Align)
SOURCE = Path(r"C:\path\to\experiment.nd2")

# Positions to crop (must match bbox/Pos{n}.csv files)
POSITIONS = list(range(0, 72))

FORCE_CROP = False
```

## Setup

```python
from pyama.services import crop

workspace = WORKSPACE.expanduser().resolve()
source = SOURCE.expanduser().resolve()
if not workspace.is_dir():
    raise FileNotFoundError(f"Workspace not found: {workspace}")
if not source.is_file():
    raise FileNotFoundError(f"Source not found: {source}")

positions = sorted({int(p) for p in POSITIONS})
if not positions:
    raise ValueError("POSITIONS is empty")

print(f"Workspace: {workspace}")
print(f"Source: {source}")
print(f"Positions ({len(positions)}): {positions}")
```

## Crop

```python
result = crop.run_crop(
    workspace=workspace,
    source=source,
    positions=positions,
    force=FORCE_CROP,
    on_progress=print,
)
for written in result.written:
    print(crop.format_written_crop_message(written))
if result.skipped_existing:
    print(f"Skipped existing ROI dirs: {result.skipped_existing}")
if result.skipped_missing_bbox:
    print(f"Skipped missing bbox CSVs: {result.skipped_missing_bbox}")

print("Crop finished. Continue in analyze.ipynb.")
```
